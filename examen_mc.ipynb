Ingresamos las libreerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

ingresamos los datos y cargamos lasprimeras filas

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv'
df = pd.read_csv(url)
print(df.tail())
resumen = df.describe()
print(resumen)
```

Eliminar valores nulos

```python
df_limpio = df.query("x > 0 and y > 0 and z > 0")
print(df_limpio)
```

Volumen de piezas

```python

df['volumen'] = df_limpio['x'] * df_limpio['y'] * df_limpio['z']
df[['x', 'y', 'z', 'volumen']].head()
```

Gráfico  de cajas


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_limpio['price'], color="skyblue")

plt.title('Detección  de  outliers', fontsize=15)
plt.xlabel('Precio')
plt.show()
```

gráfico de scatter


```python
plt.figure()
plt.scatter(df["volumen"], df["price"])
plt.title("volumen y precio")
plt.xlabel("Volumen Aproximado")
plt.ylabel("Precio")
plt.show()
```

Gráfico de barras

```python
cantidad_por_cortes = df['cut'].value_counts()
sns.barplot(x=cantidad_por_cortes.index, y=cantidad_por_cortes.values, palette="viridis")
plt.title('Cantidad Total de Diamantes por Tipo de Corte', fontsize=15)
plt.xlabel('Tipo de Corte (Cut)', fontsize=12)
plt.ylabel('Cantidad de Piezas', fontsize=12)
```
